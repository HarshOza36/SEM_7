# Experiment No :12 - Image Segmentation

```python
import numpy as np
import cv2 as cv2
from google.colab.patches import cv2_imshow # for image display
from skimage import io
import matplotlib.pylab as plt
```

```python
url = "https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg"
#url = "https://homepages.inf.ed.ac.uk/rbf/HIPR2/images/fce5noi5.gif"
#url = "http://eeweb.poly.edu/~yao/EL5123/image/lena_gray.bmp"
image = io.imread(url)
gray_image =image
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
m,n = gray_image.shape
img=gray_image
ct= np.empty(shape=(m,n),dtype='uint8')
```

# Image Segmentation using Sobel Filter

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=3)
print('Sobel X')
cv2_imshow(sobelx)
print('Sobel Y')
cv2_imshow(sobely)
```

```python
img_sobel = sobelx + sobely
print("Sobel Magnitude")
cv2_imshow(img_sobel)
```

# Image Segmentation using Prewitt Filtering

```python
#prewitt
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
img_prewittx = cv2.filter2D(img, -1, kernelx)
img_prewitty = cv2.filter2D(img, -1, kernely)
print('Prewitt X')
cv2_imshow(img_prewittx)
print('Prewitt Y')
cv2_imshow(img_prewitty)
```

```python
img_prewitt = img_prewittx + img_prewitty
print("Prewitt Magnitude")
cv2_imshow(img_prewitt)
```
