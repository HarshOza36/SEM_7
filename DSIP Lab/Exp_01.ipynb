# **Question 1**

```
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```
from scipy import signal
x = np.linspace(0, 10, 20, endpoint=False)
# The line above generate x axis between 0 to 10 20 points
y = np.sin(x*2*3.14/10.0)
# The line below is used to plot and show the sine wave signal
plt.plot(x,y)
plt.show()
```

```
f = signal.resample(y, 10)
xnew = np.linspace(0, 10, 10, endpoint=False)
plt.plot(x, y, 'go-', xnew, f, '.-', 10, y[0], 'ro')
plt.legend(['data', 'resampled'], loc='best')
plt.show()
```

```
f = signal.resample(y, 20)
xnew = np.linspace(0, 10, 20, endpoint=False)
plt.plot(x, y, 'go-', xnew, f, '.-', 10, y[0], 'ro')
plt.legend(['data', 'resampled'], loc='best')
plt.show()
f = signal.resample(y, 24)
```

```
f = signal.resample(y, 28)
xnew = np.linspace(0, 10, 28, endpoint=False)
plt.plot(x, y, 'go-', xnew, f, '.-', 10, y[0], 'ro')
plt.legend(['data', 'resampled'], loc='best')
plt.show()
d = signal.decimate(f,1)
plt.plot(d)
```
