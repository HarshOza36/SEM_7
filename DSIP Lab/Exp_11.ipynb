# Experiment No :11 - Image Smoothing and Sharpening


```python
import numpy as np
import cv2 as cv2
from google.colab.patches import cv2_imshow # for image display
from skimage import io
import matplotlib.pylab as plt
```

```python
#url = "https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg"
url = "https://homepages.inf.ed.ac.uk/rbf/HIPR2/images/fce5noi5.gif"
#url = "http://eeweb.poly.edu/~yao/EL5123/image/lena_gray.bmp"
image = io.imread(url)
gray_image =image
#gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
m,n = gray_image.shape
img=gray_image
ct= np.empty(shape=(m,n),dtype='uint8')
```

# Image Smoothing using Mean Filter

```python
kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(img,-1,kernel)
print("Original")
cv2_imshow(img)
print("Averaging")
cv2_imshow(dst)
```

# Image Smoothing using Median Filter

```python
url = "http://www.fit.vutbr.cz/~vasicek/imagedb/img_corrupted/impnoise_005/108073.png"
#url = "http://eeweb.poly.edu/~yao/EL5123/image/lena_gray.bmp"
image = io.imread(url)
gray_image =image
#gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
img=gray_image


median = cv2.medianBlur(img, 3)
cv2_imshow(median)
```

# Sharpening Filter : High Pass Filter

```python
#url = "https://homepages.inf.ed.ac.uk/rbf/HIPR2/images/fce5noi5.gif"
url = "http://eeweb.poly.edu/~yao/EL5123/image/lena_gray.bmp"
image = io.imread(url)
gray_image =image
img = gray_image
#gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
kernel = np.array([[-1, -1, -1],
                       [-1, 8, -1],
                       [-1, -1, -1]])
kernel = kernel/9
hpf1 = cv2.filter2D(img,-1,kernel)
print("High Pass Filter Output")
cv2_imshow(hpf1)
```

```python
kernelhbf = np.array([[-1, -1, -1],
                       [-1, 9, -1],
                       [-1, -1, -1]])
hpf1 = cv2.filter2D(img,-1,kernelhbf)
print("High Boost Filter Output")
cv2_imshow(hpf1)
```
